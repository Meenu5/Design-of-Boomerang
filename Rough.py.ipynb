```python
import numpy as np

# Generating testbed for 'total' number of bandit settings with 'k' arms each

def doTestbed(k, total) :
    # Initialization
    true_arm_index = np.zeros([total,1],int)
    qstar_mu = np.zeros([total,k])

    for i in range(total) :
        qstar_mu[i,:] = np.random.randn(1,k)        # true reward mean sampled from normal distribution with mean 0, unit variance
        true_arm_index[i] = qstar_mu[i,:].argmax(0) # true best arms for k bandits    

    return qstar_mu, true_arm_index
```

```python
import numpy as np
import math

# k - arms, total - number of bandit settings, runs - steps, c - parameter, qstar_mu - mean of true reward, true_arm_index - true best arm
def doUcb(k, total, runs, c, qstar_mu, true_arm_index) :

    # Initialization
    q_value = np.zeros([total,k])
    n_arms = np.ones([total,k])

    avg_reward = []
    optimal_action_norm = []

    for i in range(runs-k+1) :
        reward_sum = 0
        optimal_action = 0

        for j in range(total) :

            if i == 0 :
                q_value = np.random.randn(total,k) + qstar_mu   # sample all the arms for the first time
                a = list(np.mean(q_value,0))
                for value in a :
                    avg_reward.append(value)
                break

            confidence_band = q_value[j,:] + c*(math.log(i+k) / n_arms[j,:]) ** 0.5
            arm_index = confidence_band.argmax(0)               # Pull the arm that has maximum value of reward band

            if arm_index  == true_arm_index[j] :
                optimal_action += 1

            arm_reward = np.random.randn() + qstar_mu[j,arm_index]  # reward of sampled arm
            n_arms[j,arm_index] += 1
            q_value[j,arm_index] += (arm_reward - q_value[j,arm_index]) / n_arms[j,arm_index]   # incrementally updating action value
            reward_sum += arm_reward

        avg_reward.append(reward_sum / total)               # average reward over 2000 bandit settings
        optimal_action_norm.append(optimal_action / total)  # optimal action %
    del avg_reward[-1]
    return avg_reward, optimal_action_norm
```

```python
def doT0TjTransformation(Lamda,theta_pitch,beta) :
    Tj = np.zeros([3,3])
    Tj[0,0] = cos(theta_pitch)*sin(Lamda)+sin(theta_pitch)*sin(beta)*cos(Lamda)
    Tj[0,1] = -cos(theta_pitch)*cos(Lamda)+sin(theta_pitch)*sin(beta)*sin(Lamda)
    Tj[0,2] = -sin(theta_pitch)*cos(beta)
    Tj[1,0] = -cos(beta)*cos(Lamda)
    Tj[1,1] = cos(beta)*sin(Lamda)
    Tj[1,2] = sin(beta)
    Tj[2,0] = sin(theta_pitch)*sin(Lamda)-cos(theta_pitch)*sin(beta)*cos(Lamda)
    Tj[2,1] = -sin(theta_pitch)*cos(Lamda)-cos(theta_pitch)*sin(beta)*sin(Lamda)
    Tj[2,2] = cos(theta_pitch)*cos(beta)
    return Tj
```
